```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.svm import SVR
import statsmodels.formula.api as smf
```

```python
data = pd.read_csv("C:/Users/3PX00LA_RS4/Documents/ATOS challenge/Demanda/DemandDataCSV.csv")
```

```python
data.head()
```

```python
lm = smf.ols(formula="Q~Month+Year+Pob+Rain+AvDHI+MaxDHI+MinWind+AvWind+MaxWind+MinRH+AvRH+MaxRH+MinT+AvT+MaxT", data= data).fit()
print("Parametros de función")
print(str(lm.params))
print("-----------------------------------------------------------------------------")

print("P-values") 
print(str(lm.pvalues))
print("-----------------------------------------------------------------------------")

print("R2 adj") 
print(str(lm.rsquared_adj))
print("-----------------------------------------------------------------------------")

lm = smf.ols(formula="Month~Year+Pob+Rain+AvDHI+MaxDHI+MinWind+AvWind+MaxWind+MinRH+AvRH+MaxRH+MinT+AvT+MaxT", data=data).fit()
rsquared = lm.rsquared
VIF = rsquared = 1/(1-rsquared)
print("VIF value") 
print(str(VIF))
print("-----------------------------------------------------------------------------")

Q_pred = lm.predict(data[["Month","Year","Pob","Rain","AvDHI","MaxDHI","MinWind","AvWind","MaxWind","MinRH","AvRH","MaxRH","MinT","AvT","MaxT"]])
print("Q predecida utilizando parametros de función") 
print(str(Q_pred))
print("-----------------------------------------------------------------------------")

SSD = sum((data["Q"]-Q_pred)**2)
print ("SSD") 
print(str(SSD))
print("-----------------------------------------------------------------------------")

RSE = np.sqrt(SSD/(len(data)-26))
print("RSE") 
print(str(RSE))
print("-----------------------------------------------------------------------------")

q_m = np.mean(data["Q"])
print("Q promedio ") 
print(str(q_m))
print("-----------------------------------------------------------------------------")

error = RSE / q_m
print("Porcentaje de error") 
print(str(error))
print("-----------------------------------------------------------------------------")

lm.summary()
```
