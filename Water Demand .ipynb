## Month + Year + Pob + Rain + INFLA

```python
lm5 = smf.ols(formula="Q~Month+Year+Pob+Rain+INFLA", data= data).fit()
print("Parametros de función")
print(str(lm5.params))
print("-----------------------------------------------------------------------------")

print("P-values") 
print(str(lm5.pvalues))
print("-----------------------------------------------------------------------------")

print("R2 adj") 
print(str(lm5.rsquared_adj))
print("-----------------------------------------------------------------------------")

lm_INFLA = smf.ols(formula="INFLA~Month+Year+Pob+Rain", data=data).fit()
rsquared_INFLA = lm_INFLA.rsquared
VIF = rsquared_INFLA = 1/(1-rsquared_INFLA)
print("VIF value ") 
print(str(VIF))
print("-----------------------------------------------------------------------------")

Q_pred = lm5.predict(data[["Month", "Year", "Pob", "Rain", "INFLA"]])
print("Q predecida utilizando parametros de función") 
print(str(Q_pred))
print("-----------------------------------------------------------------------------")

SSD = sum((data["Q"]-Q_pred)**2)
print ("SSD") 
print(str(SSD))
print("-----------------------------------------------------------------------------")

RSE = np.sqrt(SSD/(len(data)-6))
print("RSE") 
print(str(RSE))
print("-----------------------------------------------------------------------------")

q_m = np.mean(data["Q"])
print("Q promedio ") 
print(str(q_m))
print("-----------------------------------------------------------------------------")

error = RSE / q_m
print("Porcentaje de error") 
print(str(error))
print("-----------------------------------------------------------------------------")

lm5.summary()
```
